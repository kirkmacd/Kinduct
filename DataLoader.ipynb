__The downloaded csv file from Kaggle contained so rows of data of varying length__  
__This function parses through each line of the csv and check the number for columns__
__If different from the expected, the line gets written to a seperate csv file for further investigation__

```python
#Function to clean file before loading into df
def file_cleaner(path, num_cols):
    with open(path) as f: #open csv file
        for i,line in enumerate(f):
            if len(line.split(',')) != num_cols: #check how many columns are in the line
                print('Error found trying to load row{}, writing row to goaliedata_error.csv'.format(i))
                with open("goaliedata_error.csv", "a", newline='\n') as file: #write to new file if issue found
                    file.write(line.split()[0]+"\n")
            else:
                with open("goaliedata_clean.csv", "a", newline='\n') as file: #write to new file if no issue found
                    file.write(line.split()[0]+"\n")
```

```python
file_cleaner("Goalies.csv", 23)
```

```python
#read Goalies file after downloading and specify what columns to load
data = pd.read_csv('goaliedata_clean.csv', 
                   usecols = ['playerID','year','tmID','GP','Min','W','L', 'T/OL','ENG','SHO','GA','SA']
                  )
```
