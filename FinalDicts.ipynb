```python
import pandas as pd
```

# Calculating most_goals_stopped {‘playerID’: playerID, ‘goals_stopped’: goals_stopped}  
- Description: calculate goals stopped per player, then take the player with the max goals stopped and put the details in the dictionary

```python
#read goaliedata_clean file after parsing file, specifying which columns to load
data = pd.read_csv('goaliedata_clean.csv', 
                   usecols = ['playerID','year','tmID','GP','Min','W','L', 'T/OL','ENG','SHO','GA','SA']
                  )
```

```python
#Group by playerID to get career totals
playerdata = (data.groupby('playerID')
              .agg({'GA':'sum',
                    'SA':'sum',
                    'Min' : 'sum'
              })
             )

#Calculate goals_stopped and efficiency
playerdata['goals_stopped'] = playerdata['SA'] - playerdata['GA']
playerdata['efficiency'] = playerdata['goals_stopped']/playerdata['Min']

#Remove unneeded columns
goals_stopped_df = playerdata.drop(columns = ['GA','SA','Min','efficiency'])

#Sort df to find who has the most goals stopped
goals_stopped_df = (goals_stopped_df
                    .sort_values(by=['goals_stopped'],
                                 ascending = False)
                    .reset_index()
                   )

#Take top entry from df and turn it into a dictionary
goals_stopped_dict = (goals_stopped_df
                      .iloc[0]
                      .to_dict()
                     )
```

# Calculating most_efficient_player: {‘playerID’: playerID, ‘efficiency’: goals_stopped / minutes played}
- Description: calculate the goals stopped per minutes of play for each player, then take the  player with the max efficiency just calculated and put the details in the dictionary

```python
#Remove unneeded columns
efficiency_df = playerdata.drop(columns = ['GA','SA','Min','goals_stopped'])

#Sort df to find who has the most goals stopped
efficiency_df = (efficiency_df
                 .sort_values(by=['efficiency'], 
                              ascending = False)
                 .reset_index())

#Take top entry from df and turn it into a dictionary
efficiency_dict = (efficiency_df
                   .iloc[0]
                   .to_dict())
```
